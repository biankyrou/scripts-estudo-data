# Primeira etapa
Essa etapa consiste em extrair do dataset inicial um segundo dataset que comporta determinada variável binária. <br>
Ex: um dataset de clientes que cancelaram o plano / outro dataset de clientes que não cancelaram o plano

```python
def dividir_grupos(df, target_col):
    if df[target_col].dtype == 'int64':
        grupo_1 = df[df[target_col] == 1]
        grupo_0 = df[df[target_col] == 0]
        
    else:
        grupo_1 = df[df[target_col] == 'Yes']
        grupo_0 = df[df[target_col] == 'No']
    
    return grupo_1, grupo_0
```
